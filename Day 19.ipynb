```python
import re
def day_19_part1():
    new_rules = dict()
    rule_re = re.compile(r"^(\d+): (?:(?:(\d+) (\d+) \| (\d+) (\d+))|(?:(\d+) (\d+))|(?:(\d+) \| (\d+))|(?:\"([ab])\"))$")
    #rule_re = re.compile("(\d+): (?:(?:(\d+) ?(\d+)?(?: \| (\d+) (\d+))?)|(?:\"([ab])\"))")

    for x in day_19_rules:
        rule_num, r = x.split(":")

        sp = r.split("|")
        sp = [x.strip(' ').split(" ") for x in sp]

        #if there is an alternation
        if len(sp) == 2:
            rule = "((" + ")(".join(sp[0]) + ")|(" + ")(".join(sp[1]) + "))"
        else:
            rule = ("((" + ")(".join(sp[0]) + "))").replace("\"", "")

        new_rules[int(rule_num)] = rule

    num_re = re.compile("\d+")

    old_str = new_rules[0]
    done = False
    while not done:
        matches = num_re.findall(old_str)
        if matches:
            new_str = ""
            for x in matches:
                new_str = old_str.replace("(" + x + ")", new_rules[int(x)])


            old_str = new_str.replace("(a)", "a").replace("(b)", "b")
        else:
            done = True

    for k, v in replacements:
        old_str = old_str.replace(k, v)
    the_re = re.compile("^" + old_str + "$")
    #print(old_str)
    return(sum(1 if the_re.match(x) else 0 for x in day_19_input))

print(day_19_part1())#should be 226
%timeit day_19_part1()
```
