```python
import numpy as np
import itertools

day_20_input[0][0][5:]

# intab = ".#"
# outtab = "01"
# trantab = str.maketrans(intab, outtab)

# intab = ".#"
# negouttab = "10"
# negtrantab = str.maketrans(intab, negouttab)
# trantab
def to_num(c):
    if c =="#":
        return 1
    else:
        return 0

def get_scores(img):
    bin_to_int = 2**np.arange(10)[::-1]
    t = img[0].dot(bin_to_int)
    r = img[:, -1].dot(bin_to_int)
    b = img[9].dot(bin_to_int)
    l = img[:, 0].dot(bin_to_int)

    return [t, r, b, l]
    
class Tile():
    def __init__(self, t):
        #print(t)
        self.tile_id = int(t[0][5:])
        
        self.image = np.array([[to_num(c) for c in s] for s in t[1]], int)
        rotations = [np.rot90(self.image, k=i) for i in range(4)] + [np.rot90(np.flipud(self.image), k = i) for i in range(4)]
        self.possible = [get_scores(r) for r in rotations]
        

tiles = [Tile(x) for x in day_20_input]



#find all tiles / orientations that can fill in the specified position in the grid
#This fills in left to right, top to bottom so we also pass in the values that must appear
#to the left or above.  Finally put in the list of tiles left to use
def get_possible_tiles(pos, left, top, tiles_rem):
    px, py = pos
#     px = pos[0]
#     py = pos[1]
    
    lch = set()
    tch = set()
    
    #get all possible tile / rotations
    all_possibilities = itertools.product(tiles_rem, range(8))
    allowed = []
    
    for tile, rot in all_possibilities:
        if px > 0 and tile.possible[rot][3] != left:
            continue
        if py > 0 and tile.possible[rot][0] != top:
            continue
            
        allowed.append((tile, rot))
        
    return allowed
    
def inc_pos(pos):
    px, py = pos
    if px == 11:
        px = 0
        py += 1
    else:
        px += 1
        
    return (px, py)

def get_constraints(grid, pos):
    px, py = pos
    l = 0
    t = 0
    
    if px > 0:
        l = grid[py][px - 1][0].possible[grid[py][px - 1][1]][1]
    
    if py > 0:
        t = grid[py - 1][px][0].possible[grid[py - 1][px][1]][2]
        
    return l, t
    

def solve(grid, tiles, tiles_rem, pos, l, t):
    #print("Solving for position", pos, "with constraints", l, t, "and", len(tiles_rem), "tiles remaining")
    px, py = pos
    #Check that there are no more tiles left to place
    #if no tiles are left, we are done
    if not tiles_rem:
        return True
    
    #Grab all choices for the next tile
    #print(pos, l, t, tiles_rem)
    choices = get_possible_tiles(pos, l, t, tiles_rem)
    #print(len(choices), "choices meet the constraint at", pos)
    #print(choices)
    while choices:
        #Take the next choice from the list, and assign it into the grid
        ch = choices.pop()
        #print("Taking choice", ch)
        grid[py][px] = ch
        #print(grid[0])
        new_tiles_rem = [t for t in tiles_rem if t != ch[0]]

        #print(len(new_tiles_rem), pos, l, ch[0].possible[ch[1]][3], t, ch[0].possible[ch[1]][0])
        
        new_pos = inc_pos(pos)
        
        ll, tt = get_constraints(grid, new_pos)
        result = solve(grid, tiles, new_tiles_rem, inc_pos(pos), ll, tt)
        
        if result:
            
            return True
        else:
            grid[py][px] = -1
            
    return False

def day_20_part1():
    global good_grid
    width = 12
    height = 12
    grid = [[-1 for x in range(width)] for y in range(height)]
    tiles_rem = [t for t in tiles]
    x = solve(grid, tiles, tiles_rem, (0, 0), 0, 0)
    
    
    good_grid = grid
    return grid[0][0][0].tile_id * grid[11][0][0].tile_id * grid[0][11][0].tile_id * grid[11][11][0].tile_id

print(day_20_part1()) #part 1 should be 15670959891893
```

```python
import numpy as np

def day_20_part2():
    new_grid = []

    width = 12
    height = 12

    for y in range(height):
        to_join = []
        for x in range(width):
            #get the smaller image
            t = good_grid[y][x][0].image[1:9, 1:9]

            #get the correct rotation
            if good_grid[y][x][1] <= 3:
                d = np.rot90(t, k=good_grid[y][x][1])
            else:
                d = np.rot90(np.flipud(t), k=good_grid[y][x][1] - 4)

            #The individual tiles will be joined all at once with np.concatenate
            to_join.append(d)

        #axis 1 is horizontal
        new_grid.append(np.concatenate(to_join, axis=1))

    #axis 0 is vertical
    new_grid = np.concatenate(new_grid, axis = 0)
    #new_grid is now the complete image, which may still need rotations

    sea_monster ="""                  # \n#    ##    ##    ###\n #  #  #  #  #  #   """
    sea_monster = np.array([[to_num(c) for c in s] for s in sea_monster.split("\n")], int)

    s_shape = np.shape(sea_monster)

    rotations = [np.rot90(new_grid, k=i) for i in range(4)] + [np.rot90(np.flipud(new_grid), k = i) for i in range(4)]

    for r in rotations:
        monster = 0
        for y in range(96 - s_shape[0]):
            for x in range(96 - s_shape[1]):
                #multiply the sea monster by the region starting at x, y
                sub_m = r[y:y+s_shape[0], x:x+s_shape[1]]
                score = sum(sum(np.multiply(sub_m, sea_monster)))
                if score == 15:
                    monster += 1
        if monster != 0:
            break

    return sum(sum(new_grid)) - (sum(sum(sea_monster)) * monster)
    
print(day_20_part2()) #should be 1964
```
