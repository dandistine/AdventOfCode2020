```python
def score_deck(deck):
    return sum([i * deck[j] for i, j in zip(range(len(deck), 0, -1), range(len(deck)))])

def day_22_part1():
    p1_deck = [x for x in player_1_start]
    p2_deck = [x for x in player_2_start]

    while p1_deck and p2_deck:
        p1, p1_deck = p1_deck[0], p1_deck[1:]
        p2, p2_deck = p2_deck[0], p2_deck[1:]

        if p1 > p2:
            p1_deck.append(p1)
            p1_deck.append(p2)
        else:
            p2_deck.append(p2)
            p2_deck.append(p1)

    return score_deck(p1_deck) + score_deck(p2_deck)

print(day_22_part1()) #should be 32162
```

```python
#return positive for P1 win, negative for P2
def combat(p1_deck, p2_deck):
    rounds = set()
    
    while p1_deck and p2_deck:
        round_t = (tuple(p1_deck), tuple(p2_deck))
        if round_t in rounds:
            return 1
        else:
            rounds.add(round_t)
        
        p1 = p1_deck[0]
        p2 = p2_deck[0]
        
        if p1 < len(p1_deck) and p2 < len(p2_deck):
            winner = combat(p1_deck[1:p1+1], p2_deck[1:p2+1])
        else:
            winner = p1 - p2

        if winner > 0:
            p1_deck = (p1_deck + [p1, p2])[1:]
            p2_deck = p2_deck[1:]

        else:
            p1_deck = p1_deck[1:]
            p2_deck = (p2_deck + [p2, p1])[1:]

    if p1_deck:
        return score_deck(p1_deck)
    else:
        return -1 * score_deck(p2_deck)

def day_22_part2():
    p1_deck = [x for x in player_1_start]
    p2_deck = [x for x in player_2_start]
    
    score = combat(p1_deck, p2_deck)
    
    return abs(score)
    
print(day_22_part2()) #should be 32534
%timeit day_22_part2()
```
