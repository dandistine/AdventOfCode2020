```python
def day_12_part1():
    rotate = {'L' : {'N' : 'WSEN', 'W' : 'SENW', 'S' : 'ENWS', 'E' : 'NWSE'},
              'R' : {'N' : 'ESWN', 'W' : 'NESW', 'S' : 'WNES', 'E' : 'SWNE'}}

    dist = {'N' : 0, 'E' : 0, 'S' : 0, 'W' : 0}
    direc = 'E'

    for x in day_12_input:
        if x[0] in 'NESW':
            dist[x[0]] = dist[x[0]] + x[1]
        elif x[0] in 'LR':
            direc = rotate[x[0]][direc][x[1] - 1]
        else:
            dist[direc] = dist[direc] + x[1]

    return abs(dist['N'] - dist['S']) + abs(dist['E'] - dist['W'])

def day_12_part2():
    waypoint = [1, 10]
    pos = [0, 0]
    
    for x in day_12_input:
        if x[0] == 'N':
            waypoint[0] += x[1]
        elif x[0] == 'S':
            waypoint[0] -= x[1]
        elif x[0] == 'E':
            waypoint[1] += x[1]
        elif x[0] == 'W':
            waypoint[1] -= x[1]
        elif x[0] == 'R':
            for i in range(x[1]):
                waypoint = [waypoint[1] * -1, waypoint[0]]
        elif x[0] == 'L':
            for i in range(x[1]):
                waypoint = [waypoint[1], waypoint[0] * -1]
        else: #move forward
            pos[0] = x[1] * waypoint[0] + pos[0]
            pos[1] = x[1] * waypoint[1] + pos[1]
            
    return(abs(pos[0]) + abs(pos[1]))
            

print(day_12_part1()) #should be 2297
print(day_12_part2()) # should be 89984
%timeit day_12_part1()
%timeit day_12_part2()
```
