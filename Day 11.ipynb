```python
from copy import deepcopy
from collections import Counter


def day_11_part1():
    start = deepcopy(day_11_input)
    temp = deepcopy(day_11_input)
    #create a fake difference so the loop will run
    temp[0][0] = 'x'

    WIDTH = len(day_11_input[0])
    HEIGHT = len(day_11_input)

    def next_state(x, y, seats):
        c = 0

        xstart = x
        xend = x + 1
        ystart = y
        yend = y + 1

        #floor never changes
        if seats[y][x] == '.':
             return seats[y][x]

        if(x > 0):
            xstart = x - 1
        if(x < WIDTH - 1):
            xend = x + 2
        if(y > 0):
            ystart = y - 1
        if(y < HEIGHT - 1):
            yend = y + 2

        if(seats[y][x] == '#'):
            c = - 1


        for i in range(xstart, xend):
            for j in range(ystart, yend):
                if(start[j][i] == '#'):
                    c += 1

        if seats[y][x] == 'L' and c == 0:
            return '#'

        if seats[y][x] == '#' and c >= 4:
            return 'L'

        return seats[y][x]

    while start != temp:
        for i in range(WIDTH):
            for j in range(HEIGHT):
                temp[j][i] = next_state(i, j, start)

        start, temp = temp, start

    seat_count = 0
    for x in start:
        for y in x:
            if y == '#':
                seat_count += 1
    return seat_count


def day_11_part2():
    start = deepcopy(day_11_input)
    temp = deepcopy(day_11_input)
    #create a fake difference so the loop will run
    temp[0][0] = 'x'


    WIDTH = len(day_11_input[0])
    HEIGHT = len(day_11_input)

    def next_state_p2(x, y, seats):
        c = 0

        if seats[y][x] == '.':
            return '.'

        i = x
        j = y - 1
        #check vertical up
        while j >= 0:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            j -= 1

        i = x + 1
        j = y - 1
        #check diagonal up and right
        while j >= 0 and i <= WIDTH - 1:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            j -= 1
            i += 1

        i = x + 1
        j = y
        #check to the right
        while i <= WIDTH - 1:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            i += 1

        i = x + 1
        j = y + 1
        #check diagonal down and right
        while i <= WIDTH - 1 and j <= HEIGHT - 1:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            i += 1
            j += 1

        i = x
        j = y + 1
        #check down
        while j <= HEIGHT - 1:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            j += 1

        i = x - 1
        j = y + 1
        #check diagonal down and left
        while i >= 0 and j <= HEIGHT - 1:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            j += 1
            i -= 1

        i = x - 1
        j = y
        #check left
        while i >= 0:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            i -= 1

        i = x - 1
        j = y - 1
        #check diagonal up and left
        while i >=0 and j >= 0:
            if seats[j][i] != '.':
                if seats[j][i] == '#':
                    c += 1
                break
            i -= 1
            j -= 1

        if seats[y][x] == 'L' and c == 0:
            return '#'

        if seats[y][x] == '#' and c >= 5:
            return 'L'

        return seats[y][x]

    while start != temp:
        for i in range(WIDTH):
            for j in range(HEIGHT):
                temp[j][i] = next_state_p2(i, j, start)

        start, temp = temp, start

    seat_count = 0
    for x in start:
        for y in x:
            if y == '#':
                seat_count += 1
    return seat_count

print(day_11_part1()) # should be 2263
print(day_11_part2()) # should be 2002
```

```python
from copy import deepcopy

def make_adj(width, height):
    adj = []
    
    for y in range(height):
        for x in range(width):
            xs = x
            xe = x + 1
            ys = y
            ye = y + 1
    
            if(x > 0):          xs = x - 1
            if(x < width - 1):  xe = x + 2
            if(y > 0):          ys = y - 1
            if(y < height - 1): ye = y + 2
            
            adj.append(sorted([j * width + i for i in range(xs, xe) for j in range(ys, ye) if not (i == x and j == y)]))
    return adj

def day_11_part1_():
    start = deepcopy(day_11_input_new)
    temp = []
    
    WIDTH = len(day_11_input[0])
    HEIGHT = len(day_11_input)

    adj = make_adj(WIDTH, HEIGHT)
    
    occ_lut =   [2, 2, 2, 2, 1, 1, 1, 1, 1]
    unocc_lut = [2, 1, 1, 1, 1, 1, 1, 1, 1]
    floor_lut = [0, 0, 0, 0, 0, 0, 0, 0, 0]
    lut = [floor_lut, unocc_lut, occ_lut]

    while temp != start:
        temp = [lut[i][sum([start[q] == 2 for q in a])] if i != 0 else 0 for i, a in zip(start, adj)]
        start, temp = temp, start
        
    return sum([i == 2 for i in start])

print(day_11_part1_())
```
